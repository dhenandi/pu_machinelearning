```python
import PIL.Image



sample_file_1 = PIL.Image.open('shogun.jpeg')
sample_file_2 = PIL.Image.open('mobil1.jpeg')
```

```python
import google.generativeai as genai
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Choose a Gemini model.
model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest")

# Create a prompt.
prompt = "You are developing an automatic parking gate that receive picture and recognizing plate number. please recognize plate number input from sample picture with a return of JSON. JSON format as below: { 'plat_no': 'B 1234 ABC', 'plat_expire': '01-22', 'vehicle': 'car', 'vehicle_type': 'sedan', 'color': 'red', 'gate_open': 'YYYY-MM-DD hh.mm.ss', 'gate_closed': 'N/A'} gate_open is current timestamp user enter the gate, for YYYY-MM-DD hh.mm.ss"

response = model.generate_content([sample_file_1, prompt])
Markdown(response.text)
```
